```python
import re

def find_emojis_emoticons(text):
    # Define regex patterns for emojis and emoticons
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emoticons
                               u"\U0001F300-\U0001F5FF"  # Symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # Transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # Flags (iOS)
                               u"\U00002600-\U000027BF"  # Miscellaneous symbols
                               u"\U0001f300-\U0001f64F"  # Additional emoticons
                               u"\U0001f680-\U0001f6FF"  # Additional transport & map symbols
                               u"\U0001f1e0-\U0001f1ff"  # Additional flags
                               u"\U00002500-\U00002BEF"  # Chinese characters
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    emoticon_pattern = re.compile(u'[\U0001F600-\U0001F64F\u263a-\U0001f645]')

    # Find emojis and emoticons in the text
    emojis = emoji_pattern.findall(text)
    emoticons = emoticon_pattern.findall(text)

    return emojis, emoticons

# Function to check if a row contains emojis or emoticons in two columns
def contains_emojis_emoticons_in_columns(row, column1, column2):
    text1 = row[column1]
    text2 = row[column2]
    emojis1, emoticons1 = find_emojis_emoticons(text1)
    emojis2, emoticons2 = find_emojis_emoticons(text2)
    return bool(emojis1) or bool(emoticons1) or bool(emojis2) or bool(emoticons2)

filtered_df = train_df[train_df.apply(lambda row: contains_emojis_emoticons_in_columns(row, 'English', 'Hindi'), axis=1)]

print(len(filtered_df))
print("\nRows containing emojis or emoticons:")
print(filtered_df)
```

```python
import re

def remove_emojis_emoticons(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F" 
                               u"\U0001F300-\U0001F5FF" 
                               u"\U0001F680-\U0001F6FF"  
                               u"\U0001F1E0-\U0001F1FF"  
                               u"\U00002600-\U000027BF"  
                               u"\U0001f300-\U0001f64F"  
                               u"\U0001f680-\U0001f6FF"  
                               u"\U0001f1e0-\U0001f1ff" 
                               u"\U00002500-\U00002BEF" 
                               u"\U00002702-\U000027B0" 
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    emoticon_pattern = re.compile(u'[\U0001F600-\U0001F64F\u263a-\U0001f645]')

    text_without_emojis = emoji_pattern.sub(r'', text)
    text_without_emoticons = emoticon_pattern.sub(r'', text_without_emojis)

    return text_without_emoticons

train_df['English'] = train_df['English'].apply(remove_emojis_emoticons)
train_df['Hindi'] = train_df['Hindi'].apply(remove_emojis_emoticons)
```

```python
from ilstokenizer import tokenizer
from joblib import Parallel, delayed

num_cores = 4

def process_text(text):
    return tokenizer.tokenize(text.lower())

train_df['English_Processed'] = Parallel(n_jobs=num_cores)(delayed(process_text)(text) for text in train_df['English'])
train_df['Hindi_Processed'] = Parallel(n_jobs=num_cores)(delayed(process_text)(text) for text in train_df['Hindi'])
```

```python
train_df.head()
```

```python
val_df['English_Processed'] = Parallel(n_jobs=num_cores)(delayed(process_text)(text) for text in val_df['English'])
val_df['Hindi_Processed'] = Parallel(n_jobs=num_cores)(delayed(process_text)(text) for text in val_df['Hindi'])


test_df['English_Processed'] = Parallel(n_jobs=num_cores)(delayed(process_text)(text) for text in test_df['English'])
test_df['Hindi_Processed'] = Parallel(n_jobs=num_cores)(delayed(process_text)(text) for text in test_df['Hindi'])
```

```python
train_df['English_Processed'].to_csv('train.src.tkn', header=None, index=None, sep=' ')
train_df['Hindi_Processed'].to_csv('train.tgt.tkn', header=None, index=None, sep=' ')

val_df['English_Processed'].to_csv('valid.src.tkn', header=None, index=None, sep=' ')
val_df['Hindi_Processed'].to_csv('valid.tgt.tkn', header=None, index=None, sep=' ')

test_df['English_Processed'].to_csv('test.src.tkn', header=None, index=None, sep=' ')
test_df['Hindi_Processed'].to_csv('test.tgt.tkn', header=None, index=None, sep=' ')
```

```python
frames = [train_df['English_Processed'], train_df['Hindi_Processed']]
result = pd.concat(frames)
train_file = 'train.all'
result.to_csv(train_file, header=None, index=None, sep=' ')
```

```python
from sacrebleu import corpus_bleu
from ilstokenizer import tokenizer

# Step 1: Read the translated output
translated_output_file = "valid.hi.translated" 
translated_output = []
with open(translated_output_file, "r") as f:
    for line in f:
        translated_output.append(line.strip())

# Step 2: Read the reference translations
reference_file = "test.hi"  
reference_translations = []
with open(reference_file, "r") as f:
    for line in f:
        reference_translations.append(line.strip())

# Step 3: Tokenize the translated output and reference translations
tokenized_output = [tokenizer.tokenize(sentence).lower() for sentence in translated_output]
tokenized_references = [[tokenizer.tokenize(sentence).lower()] for sentence in reference_translations]

# Step 4: Compute the BLEU score
bleu_score = corpus_bleu(tokenized_output, tokenized_references)
print("BLEU score:", bleu_score.score)
```
